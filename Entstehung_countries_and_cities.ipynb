```python
# Städte mit mehr als 100K Einwohner abrufen, wenn keine vorhanden, dann Hauptstadt (in Versalien) einsetzen

import requests

# API für Städte abrufen
def get_cities_by_country():
    url = "https://countriesnow.space/api/v0.1/countries/population/cities"
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status() 
        cities_data = response.json().get("data", [])
        
        country_city_map = {}
        
        for entry in cities_data:
            country = entry.get("country")
            city = entry.get("city")
            population_counts = entry.get("populationCounts", [])
            
            if not country or not city or not population_counts:
                continue

            valid_populations = []
            for pc in population_counts:
                value = pc.get("value")
                if isinstance(value, (int, float, str)):
                    try:
                        valid_populations.append(float(value))
                    except ValueError:
                        pass
            

            latest_population = max(valid_populations) if valid_populations else 0
            
            if latest_population >= 100000:
                country_city_map.setdefault(country, []).append(city)
        
        return country_city_map
    except requests.exceptions.RequestException as e:
        print(f"Fehler bei der API-Anfrage: {e}")
        return {}

empty_cities = {
    'Andorra': {'country_id': 'Q228', 'cities': []},
    'Armenia': {'country_id': 'Q399', 'cities': []},
    'Azerbaijan': {'country_id': 'Q227', 'cities': []},
    'Georgia': {'country_id': 'Q230', 'cities': []},
    'Kazakhstan': {'country_id': 'Q232', 'cities': []},
    'Kosovo': {'country_id': 'Q1246', 'cities': []},
    'Liechtenstein': {'country_id': 'Q347', 'cities': []},
    'Malta': {'country_id': 'Q233', 'cities': []},
    'Moldova': {'country_id': 'Q217', 'cities': []},
    'Monaco': {'country_id': 'Q235', 'cities': []},
    'North Macedonia': {'country_id': 'Q221', 'cities': []},
    'San Marino': {'country_id': 'Q238', 'cities': []},
    'Czech Republic': {'country_id': 'Q213', 'cities': []},
    'Turkey': {'country_id': 'Q43', 'cities': []},
    'United Kingdom of Great Britain and Northern Ireland': {'country_id': 'Q145', 'cities': []},
    'Cyprus': {'country_id': 'Q229', 'cities': []}
}

api_cities = get_cities_by_country()

capital_cities = {
    'Andorra': 'ANDORRA LA VELLA',
    'Armenia': 'YEREVAN',
    'Azerbaijan': 'BAKU',
    'Georgia': 'TBILISI',
    'Kazakhstan': 'ASTANA',
    'Kosovo': 'PRISTINA',
    'Liechtenstein': 'VADUZ',
    'Malta': 'VALLETTA',
    'Moldova': 'CHISINAU',
    'Monaco': 'MONACO',
    'North Macedonia': 'SKOPJE',
    'San Marino': 'CITY OF SAN MARINO',
    'Czech Republic': 'PRAGUE',
    'Turkey': 'ANKARA',
    'United Kingdom of Great Britain and Northern Ireland': 'LONDON',
    'Cyprus': 'NICOSIA'
}

updated_cities_data = {}

for country in empty_cities:
    if country in api_cities:
        empty_cities[country]["cities"] = api_cities[country]

    if not empty_cities[country]["cities"] and country in capital_cities:
        empty_cities[country]["cities"] = [capital_cities[country]]

    updated_cities_data[country] = empty_cities[country]

print(updated_cities_data)
```

```python
# API für Wikidata-Abfragen
def get_wikidata_entity_id(city_name):
    url = f"https://www.wikidata.org/w/api.php"
    params = {
        'action': 'wbsearchentities',
        'search': city_name,
        'language': 'en',
        'limit': 1,
        'format': 'json'
    }

    try:
        response = requests.get(url, params=params, timeout=10)
        response.raise_for_status()
        data = response.json()
        
        if data.get("search"):
            entity_id = data["search"][0]["id"]
            return entity_id
        else:
            return None  
    except requests.exceptions.RequestException as e:
        print(f"Fehler bei der Wikidata-Anfrage für {city_name}: {e}")
        return None

def add_wikidata_ids_to_cities(updated_cities_data):
    for country, country_data in updated_cities_data.items():
        for i, city in enumerate(country_data["cities"]):
            entity_id = get_wikidata_entity_id(city)
            if entity_id:
                updated_cities_data[country]["cities"][i] = (city, entity_id)
            else:
                updated_cities_data[country]["cities"][i] = (city, "Not found")
    return updated_cities_data


updated_cities_data = {'Andorra': {'country_id': 'Q228', 'cities': ['ANDORRA LA VELLA']}, 'Armenia': {'country_id': 'Q399', 'cities': ['Gyumri (Leninakan)', 'YEREVAN', 'Vanadzoz (Kirovakan)']}, 'Azerbaijan': {'country_id': 'Q227', 'cities': ['BAKU', 'Ganja', 'Sumgayit']}, 'Georgia': {'country_id': 'Q230', 'cities': ['Batumi', 'Kutaisi', 'Rustavi', 'TBILISI']}, 'Kazakhstan': {'country_id': 'Q232', 'cities': ['Aktau', 'Aktobe', 'Almaty', 'ASTANA', 'Atirau', 'Ekibastuz', 'Karaganda', 'Koktshetau', 'Kustanai', 'Kyzylorda', 'Pavlodar', 'Petropavlovsk (Severo- Kazakhstanskaya oblast)', 'Rudni', 'Semipalatinsk', 'Shimkent', 'Taldykorgan', 'Taraz', 'Temirtau', 'Uralsk', 'Ust-Kamenogorsk']}, 'Kosovo': {'country_id': 'Q1246', 'cities': ['PRISTINA']}, 'Liechtenstein': {'country_id': 'Q347', 'cities': ['VADUZ']}, 'Malta': {'country_id': 'Q233', 'cities': ['VALLETTA']}, 'Moldova': {'country_id': 'Q217', 'cities': ['CHISINAU']}, 'Monaco': {'country_id': 'Q235', 'cities': ['MONACO']}, 'North Macedonia': {'country_id': 'Q221', 'cities': ['SKOPJE']}, 'San Marino': {'country_id': 'Q238', 'cities': ['CITY OF SAN MARINO']}, 'Czech Republic': {'country_id': 'Q213', 'cities': ['Brno', 'Liberec', 'Olomouc', 'Ostrava', 'Plzen', 'PRAHA']}, 'Turkey': {'country_id': 'Q43', 'cities': ['Adana', 'Adiyaman', 'Afyonkarahisar', 'Agri', 'Akhisar', 'Aksaray', 'Alanya (Antalya)', 'ANKARA', 'Antalya', 'Aydin', 'Balikesir', 'Bandirma', 'Batman', 'Bolu', 'Bursa', 'Ceyhan', 'Cizre', 'Çanakkale', 'Çorlu', 'Çorum', 'Denizli', 'Diyarbakir', 'Düzce', 'Edirne', 'Elazig', 'Eregli', 'Erzurum', 'Eskisehir', 'Gaziantep', 'Giresun', 'Hatay', 'Inegöl', 'Iskenderun', 'Isparta', 'Istanbul', 'Izmir', 'Kahramanmaras', 'Karabük', 'Karaman', 'Kayseri', 'Kirikkale', 'Kirsehir', 'Kiziltepe', 'Kocaeli', 'Konya', 'Kütahya', 'Lüleburgaz', 'Malatya', 'Manisa', 'Mersin', 'Nazilli', 'Nigde', 'Ordu', 'Osmaniye', 'Polatli', 'Rize', 'Sakarya', 'Samsun', 'Sanliurfa', 'Siirt', 'Sivas', 'Siverek', 'Tarsus', 'Tekirdag', 'Tokat', 'Trabzon', 'Turgutlu ( Manisa)', 'Usak', 'Van', 'Yalova', 'Zonguldak']}, 'United Kingdom of Great Britain and Northern Ireland': {'country_id': 'Q145', 'cities': ['Aberdeen', 'Aberdeenshire', 'Aldershot', 'Angus', 'Basildon/North Benfleet', 'Bedford/Kempston', 'Belfast', 'Birkenhead', 'Blackburn/Darwen', 'Blackpool', 'Bournemouth', 'Brighton/Worthing/Littlehampton', 'Bristol', 'Burnley/Nelson', 'Cardiff', 'Cheltenham/Charlton Kings', 'Chesterfield/Staveley', 'Colchester', 'Coventry/Bedworth', 'Crawley', 'Dearne Valley', 'Derby', 'Doncaster', 'Dumfries & Galloway', 'Dundee', 'East Ayrshire', 'East Dunbartonshire', 'Eastbourne', 'Edinburgh', 'Exeter', 'Falkirk', 'Fife', 'Glasgow', 'Gloucester', 'Grimsby/Cleethorpes', 'Hastings/Bexhill', 'High Wycombe', 'Highland', 'Ipswich', 'Kingston-upon-Hull', 'Leicester', 'Lincoln', 'Liverpool', 'LONDON', 'Luton/Dunstable', 'Manchester', 'Mansfield', 'Milton Keynes', 'Newport', 'North Ayrshire', 'North Lanarkshire', 'Northampton', 'Norwich', 'Nottingham', 'Nuneaton', 'Oxford', 'Perth & Kinross', 'Plymouth', 'Portsmouth', 'Preston', 'Reading/Wokingham', 'Renfrewshire', 'Scottish Borders', 'Sheffield', 'Slough', 'South Ayrshire', 'South Lanarkshire', 'Southampton', 'Southend', 'Southport/Formby', 'St. Albans/Hatfield', 'Sunderland', 'Swansea', 'Swindon', 'Teesside', 'Telford', 'Thanet', 'The Medway Towns', 'The Potteries', 'Torbay', 'Tyneside', 'Warrington', 'West Lothian', 'West Midlands', 'West Yorkshire', 'Wigan', 'York']}, 'Cyprus': {'country_id': 'Q229', 'cities': ['NICOSIA']}}


updated_cities_with_ids = add_wikidata_ids_to_cities(updated_cities_data)


print(updated_cities_with_ids)
```
